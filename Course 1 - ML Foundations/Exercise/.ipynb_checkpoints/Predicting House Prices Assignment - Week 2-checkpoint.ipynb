# Question 3

```python
training_data, test_data = sales.random_split(.8,seed=0)
```

```python
my_features = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'zipcode']
advanced_features = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'zipcode', 
                     'condition', # condition of house
                     'grade', # measure of quality of construction
                     'waterfront', # waterfront property
                     'view', # type of view
                     'sqft_above', # square feet above ground
                     'sqft_basement', # square feet in basement
                     'yr_built', # the year built
                     'yr_renovated', # the year renovated
                     'lat', 
                     'long', # the lat-long of the parcel
                     'sqft_living15', # average sq.ft. of 15 nearest neighbors 
                     'sqft_lot15']
```

```python
my_features_model = tc.linear_regression.create(training_data, target="price", features=my_features, validation_set=None)
```

```python
advanced_features_model = tc.linear_regression.create(training_data, target="price", features=advanced_features, validation_set=None)
```

```python
print(my_features_model.evaluate(test_data))
```

```python
print(advanced_features_model.evaluate(test_data))
```
